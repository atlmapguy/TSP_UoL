```python
import matplotlib.pyplot as plt
import numpy as np
import itertools
import math
import pandas as pd
import io
```

```python
#calculate distance between two points

def dist_bet_two_pts(x1, y1, x2, y2):
    dist = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return dist
```

```python
#input coordinates

coord1 = (30.147059, 79.746835)
coord2 = (45.996732, 78.270042)
coord3 = (61.192810, 78.270042)
coord4 = (60.049020, 48.734177)
coord5 = (43.545752, 49.156118)
coord6 = (30.228758, 51.476793)
```

```python
coord_list = [coord1, coord2, coord3, coord4, coord5, coord6]
```

```python
#get a list of tuple index for further manipulation
index_list = []
index_pos = 0
for i in range(len(coord_list)):
    index_list.append(index_pos)
    index_pos += 1
    
#get all permutations of pts sequence
combi = list(itertools.permutations(index_list))    
```

```python
final_arr_of_dist = [[]]
final_list_of_combi = []

combi_index_counter = 0 

for i in range(len(combi)):
    
    #create list of coordinate tuple for a particular combination
    new_list = []
    tuple_index_counter = 0
    for i in range(len(combi[combi_index_counter])):
        num = (combi[combi_index_counter])[tuple_index_counter]
        new_list.append(coord_list[num])
        tuple_index_counter += 1
        
    #calculate distance between each and every point, in that order coordinates sequence
    firstpt = 0
    secpt = 1
    arr_of_dist = [[]]

    for i in range(len(new_list)-1):
        calc_dist = dist_bet_two_pts(((new_list[firstpt])[0]), ((new_list[firstpt])[1]), 
                                     ((new_list[secpt])[0]), ((new_list[secpt])[1]))
        arr_of_dist = np.append(arr_of_dist, calc_dist)
        firstpt += 1
        secpt +=1
    
    #calculate dist between first and last point
    calc_dist_return = dist_bet_two_pts(((new_list[0])[0]), ((new_list[0])[1]),
                                        ((new_list[len(new_list)-1])[0]), ((new_list[len(new_list)-1])[1]))
    
    arr_of_dist = np.append(arr_of_dist, calc_dist_return )
        
    total_dist = np.sum(arr_of_dist)
    final_arr_of_dist = np.append(final_arr_of_dist, total_dist)
    final_list_of_combi.append(new_list)
    print (total_dist, new_list)
    print('')
    
    # Creating a dataframe object from listoftuples
    d = (total_dist, new_list)
    df = pd.DataFrame(d) 
    
    combi_index_counter += 1
```
